```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
data.head(10)
```

```python
# Data visualization
# We need to encode the column 'sex'.

from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
data['Sex'] = label_encoder.fit_transform(data['Sex'])
data['Embarked'] = label_encoder.fit_transform(data['Embarked'])
data['Cabin'] = label_encoder.fit_transform(data['Cabin'])
test_data['Sex'] = label_encoder.fit_transform(test_data['Sex'])
test_data['Embarked'] = label_encoder.fit_transform(test_data['Embarked'])
test_data['Cabin'] = label_encoder.fit_transform(test_data['Cabin'])
data.head(10)
```

```python
corelation = data.corr()
corelation.style.background_gradient(cmap='coolwarm')
```

We can see that 'Sex' and 'Pclass' has pretty good correlation with survived. Out of these, 'Sex' seems to have the most correlation.

```python
# Selecting only the data we need
select_features = ['PassengerId', 'Pclass', 'Sex', 'Age',]
x = data[select_features]
y = data['Survived']
x_test = test_data[select_features]

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=1000)
model.fit(x,y)
predictions = model.predict(x_test)

# output

output = pd.DataFrame({'PassengerId':x_test.PassengerId, 'Survived': predictions})
output.to_csv('submission_4.csv', index=False)
```
